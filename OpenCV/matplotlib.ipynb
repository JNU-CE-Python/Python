```python
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.imread('dog.jpg') #BGR
plt.imshow(img1) #RGB
```

```python
plt.axis('off') #  
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) # convert
plt.imshow(img2)
```

```python
imgGray = cv2.imread('dog.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
cv2.namedWindow('image')
cv2.imshow('image', imgGray)
cv2.waitKey()
cv2.destoryALLWindows()
```
